Top 10 countries where movies/shows are produced

```python
NF.nunique().sort_values()
```

```python
top_countries =NF.groupby('country').count().sort_values('type', ascending=False)
top_countries['show_id'].head(10)
```

```python
plt.figure(figsize=(22,10))
NF["country"].value_counts()[:15].plot(kind="barh", color="Pink")
plt.title("Top 10 countries ",size=20);
```

The most popular types of movies

```python
plt.figure(figsize=(22,10))
NF["listed_in"].value_counts()[:15].plot(kind="barh", color="Green")
plt.title("Top 10 popular types of movies ",size=20);
```

Top directors **bold text**

```python
NF['director'].unique()
```

```python
top_directors =NF.groupby('director').count().sort_values('type', ascending=False)
top_directors['show_id'].head(10)
```
